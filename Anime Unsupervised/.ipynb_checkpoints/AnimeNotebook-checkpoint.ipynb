### Feature Engineering

```python
# Handling categorical data: One-hot encoding for genres and type
anime_df['genre'] = anime_df['genre'].fillna('Unknown')
genre_encoded = anime_df['genre'].str.get_dummies(sep=', ')
type_encoded = pd.get_dummies(anime_df['type'], prefix='type')

# Standardizing numerical features: members and episodes
scaler = StandardScaler()
anime_df['members_scaled'] = scaler.fit_transform(anime_df[['members']])
anime_df['episodes'] = pd.to_numeric(anime_df['episodes'], errors='coerce').fillna(1)
anime_df['episodes_scaled'] = scaler.fit_transform(anime_df[['episodes']])

# Creating interaction feature
anime_df['popularity_rating'] = anime_df['members_scaled'] * anime_df['rating']

# Concatenate all features
feature_df = pd.concat([anime_df[['rating', 'popularity_rating', 'members_scaled', 'episodes_scaled']], genre_encoded, type_encoded], axis=1)
```

##### Models

##### Collaborative filtering

```python

# Prepare the dataset
reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(train_df[['user_id', 'anime_id', 'rating']], reader)
trainset = data.build_full_trainset()

# Define and train the SVD model
svd_model = SVD()
svd_model.fit(trainset)

# Evaluate the model
results = cross_validate(svd_model, data, measures=['RMSE'], cv=3, verbose=True)
```

```python
# Create a list to hold the predictions
test_predictions = []

# Iterate over each row in your test DataFrame
for idx, row in test_df.iterrows():
    # Make a prediction for each row in the test set
    prediction = svd_model.predict(row['user_id'], row['anime_id'])
    # The 'est' attribute of the prediction object holds the estimated rating
    test_predictions.append(prediction.est)

# Optionally, add the predictions back to the test_df if you want to see them next to the respective user and anime IDs
test_df['predicted_rating'] = test_predictions

# Display the first few rows of the test dataframe to see the predicted ratings
print(test_df.head())
```

```python
# Step 1: Concatenate user_id and anime_id to form a unique ID
test_df['ID'] = test_df['user_id'].astype(str) + '_' + test_df['anime_id'].astype(str)

# Step 2: Create the submission DataFrame
submission_df = test_df[['ID', 'predicted_rating']]

# Rename columns to match the submission requirements
submission_df.columns = ['ID', 'rating']

# Round the 'rating' column to one decimal place
submission_df['rating'] = submission_df['rating'].round(1)

# Step 3: Save the DataFrame to a CSV file
submission_df.to_csv('submission.csv', index=False)

# Display the first few rows to verify
print(submission_df.head())
```

#### Content based modeling

```python
# Create a mapping from anime_id to a continuous index starting from 0
anime_indices = {anime_id: idx for idx, anime_id in enumerate(anime_df['anime_id'].unique())}

```

```python
# Convert train_df to a dictionary for faster access
user_ratings = train_df.groupby('user_id')[['anime_id', 'rating']].apply(
    lambda x: dict(zip(x['anime_id'], x['rating']))
).to_dict()

# Precompute cosine similarity matrix, ensure anime_df['anime_id'] aligns with index
cosine_sim = cosine_similarity(feature_df.drop('rating', axis=1))  # Assuming this aligns with anime_indices
```

```python
def predict_rating(user_id, anime_id):
    if anime_id not in anime_indices:
        return 3.0  # Default rating for unknown anime

    anime_idx = anime_indices[anime_id]
    try:
        user_anime_ratings = user_ratings[user_id]  # Access precomputed ratings
    except KeyError:
        return 3.0  # Default rating if user not found

    # Filter ratings and indices for known animes only
    watched_indices = [anime_indices[a_id] for a_id in user_anime_ratings if a_id in anime_indices]
    ratings = [user_anime_ratings[a_id] for a_id in user_anime_ratings if a_id in anime_indices]

    if watched_indices:
        similarities = cosine_sim[anime_idx, watched_indices]
        if np.sum(similarities) > 0:
            return np.dot(ratings, similarities) / np.sum(similarities)

    return 3.0
```
